```python
from sklearn.ensemble import RandomForestClassifier
```

# Base Credit Data - 98,4%

```python
# Coletando as variáveis do arquivo pkl
import pickle
with open('/content/drive/MyDrive/variáveis de teste e treinamento(credit, census)/credit.pkl', 'rb') as f:
    x_credit_treinamento, y_credit_treinamento, x_credit_teste, y_credit_teste = pickle.load(f)
```

```python
# Visualizando tamanho das variáveis
print(x_credit_treinamento.shape, y_credit_treinamento.shape)
print(x_credit_teste.shape, y_credit_teste.shape)
```

```python
# Declarando variável de execução das árvores de decisão
random_forest_credit = RandomForestClassifier(n_estimators=40, criterion='entropy', random_state=0)
# n_estimators=10: Esse parâmetro especifica o número de árvores de decisão a serem criadas na Random Forest.
# criterion='entropy': Esse parâmetro determina a medida usada para avaliar a qualidade de uma divisão em uma árvore de decisão.
# Está definido como 'entropia', o que significa que usará ganho de informação (entropia) para tomar decisões sobre como dividir os dados em cada nó das árvores.
# random_state=0: Esse parâmetro é usado para controlar a aleatoriedade no processo de treinamento do modelo. Ao defini-lo para um valor fixo (0 nesse caso)
```

```python
# Fazendo o treinamento
random_forest_credit.fit(x_credit_treinamento, y_credit_treinamento)
```

```python
# Realizando as previsões
previsoes_credit = random_forest_credit.predict(x_credit_teste)
```

```python
previsoes_credit
```

```python
# Importação das variáveis de accuracy
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(f'A taxa de acerto nas randons forests foram: {accuracy_score(y_credit_teste, previsoes_credit)}')
# Foi realizado testes com 100, 80, 70, 200, 100 árvores de decisão, mas com apenas 40 obtive melhor resultado
```

```python
# Gerando a matriz de confusão
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(random_forest_credit)
cm.fit(x_credit_treinamento, y_credit_treinamento)
cm.score(x_credit_teste, y_credit_teste)
```

```python
# Criando uma tabela de classificação
print(classification_report(y_credit_teste, previsoes_credit))
```
