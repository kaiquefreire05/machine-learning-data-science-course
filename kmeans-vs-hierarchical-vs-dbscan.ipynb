# K-means x Hierárquico x DBSCAN

```python
# Importações para a comparativo

from sklearn import datasets
import numpy as np
import plotly.express as px
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
```

```python
# Importando um dataset existente no sklearn

X_random, y_random = datasets.make_moons(n_samples=1500, noise = 0.09)
```

```python
# Verificando as variáveis

X_random, y_random
```

```python
# Verifincando possíveis resultados

np.unique(y_random)
```

```python
# Vendo como o gráfico está distribuído

grafico = px.scatter(x = X_random[:,0], y = X_random[:,1])
grafico.show()
```

```python
# Fazendo agrupamento usando KMeans

kmeans = KMeans(n_clusters=2, n_init=10)
rotulos = kmeans.fit_predict(X_random)
grafico = px.scatter(x = X_random[:,0], y = X_random[:, 1], color = rotulos)
grafico.show()

# kmeans 'dividiu' os registros ao meio, não sendo interessante
```

```python
# Fazendo agrupamento usando Agglomerative Clustering

hc = AgglomerativeClustering(n_clusters=2, metric='euclidean', linkage='ward')
rotulos = hc.fit_predict(X_random)
grafico = px.scatter(x = X_random[:,0], y = X_random[:, 1], color = rotulos)
grafico.show()

# agrupamento hierárquico conseguiu dividir uma parte perfeitamente, mas outra parte não foi tão boa
```

```python
# Fazendo agrupamento usando DBSCAN

dbscan = DBSCAN(eps=0.1)
rotulos = dbscan.fit_predict(X_random)
grafico = px.scatter(x = X_random[:,0], y = X_random[:, 1], color = rotulos)
grafico.show()

# nesse caso conseguiu ser dividido de melhor forma
# tudo depende da base de dados e testes
```
