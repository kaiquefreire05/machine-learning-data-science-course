# Regressão Linear Simples

## Base plano de saúde

```python
# Importações iniciais

import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.graph_objects as go

from sklearn.model_selection import train_test_split
```

```python
# Fazendo a leitura da base de dados

base_plano_saude = pd.read_csv('/content/drive/MyDrive/datasets/plano_saude.csv')
base_plano_saude
```

```python
# Pegando previsores e classe

x_plano_saude = base_plano_saude.iloc[:, 0].values # previsores
y_plano_saude = base_plano_saude.iloc[:, 1].values # classe
```

```python
x_plano_saude, y_plano_saude
```

```python
# Coeficiente de correlação

np.corrcoef(x_plano_saude, y_plano_saude)
```

```python
# Tamanho da variável antes do reshape

x_plano_saude.shape
```

```python
x_plano_saude = x_plano_saude.reshape(-1, 1)
x_plano_saude
```

```python
# Variável depois do reshape

x_plano_saude.shape
```

```python
y_plano_saude
```

```python
# Importando e fazendo o regressor

from sklearn.linear_model import LinearRegression # importação para usar a regressão linear
regressor_plano_saude = LinearRegression()
regressor_plano_saude.fit(x_plano_saude, y_plano_saude)
```

```python
# b0

regressor_plano_saude.intercept_ # valor do início da linha de regressão
```

```python
# b1

regressor_plano_saude.coef_ # indica declive da linha
```

```python
# Fazendo previsões

previsoes = regressor_plano_saude.predict(x_plano_saude)
previsoes
```

```python
# Voltar o X para o formato de vetor e plotar o gráfico

x_plano_saude.ravel() # ravel(), transforma em vetor simples
```

```python
# Gráfico para visualizar as previsões

grafico = px.scatter(x = x_plano_saude.ravel(), y = y_plano_saude)
grafico.add_scatter(x = x_plano_saude.ravel(), y = previsoes, name = 'Regressão')
grafico.show()

# a linha laranja representa o valor previsto
# os pontos azuis são os valores reais
```

```python
# Fazendo o cálculo manualmente da primeira pessoa que tem 18 anos

print(regressor_plano_saude.intercept_ + regressor_plano_saude.coef_ * 18)
# pode-se observar que o valor é o mesmo

print(regressor_plano_saude.intercept_ + regressor_plano_saude.coef_ * 40) # pessoa que não está na base, apenas teste
```

```python
regressor_plano_saude.predict([[40]]) # exatamento os mesmos valores feito automaticamente pelo algoritmo
```

```python
# Verificando a qualidade do algoritmo

regressor_plano_saude.score(x_plano_saude, y_plano_saude)  # quanto mais próximo de 1.0 melhor é o valor
```

```python
# Mostrando a distância dos valores originais para a linha da regressão linear

from yellowbrick.regressor import ResidualsPlot

visualizador = ResidualsPlot(regressor_plano_saude)
visualizador.fit(x_plano_saude, y_plano_saude)
visualizador.poof()
```

# Base preço das casas

```python
# Lendo a base de dados

base_casas = pd.read_csv('/content/drive/MyDrive/datasets/house_prices.csv')
base_casas
```

```python
# Visualizando algumas estatísticas

base_casas.describe()
```

```python
# Verificando a existência de valores nulos

base_casas.isnull().sum()
```

```python
# Correlação entre colunas

base_casas.corr()
```

```python
# Mapa de calor

fig = plt.figure(figsize=(15, 15))
sns.heatmap(base_casas.corr(), annot=True);
```

```python
# Pegando previsor e classe

x_casas = base_casas.iloc[:, 5:6].values
y_casas = base_casas.iloc[:, 2].values

x_casas, y_casas
```

```python
# Dividindo entre base de dados e teste

x_casas_treinamento, x_casas_teste, y_casas_treinamento, y_casas_teste = train_test_split(x_casas, y_casas, test_size = 0.3, random_state = 0)
```

```python
# Verificando tamanho das variáveis

print(x_casas_treinamento.shape, y_casas_treinamento.shape)
print(x_casas_teste.shape, y_casas_teste.shape)
```

```python
# Criando o regressor e treinando

regressor_simples_casas = LinearRegression() # criando regressor
regressor_simples_casas.fit(x_casas_treinamento, y_casas_treinamento) # fazendo o treinamento
```

```python
# b0

regressor_simples_casas.intercept_ # valor do início da linha de regressão
```

```python
# b1

regressor_simples_casas.coef_ # indica declive da linha
```

```python
# Verificando o score com as variáveis de treinamento

regressor_simples_casas.score(x_casas_treinamento, y_casas_treinamento) # desempenho ruim
```

```python
# Verificando o score com as variáveis de teste

regressor_simples_casas.score(x_casas_teste, y_casas_teste) # desempenho ruim
```

```python
previsoes = regressor_simples_casas.predict(x_casas_treinamento)
previsoes
```

```python
# Gráfico para visualizar as previsões

grafico = px.scatter(x = x_casas_treinamento.ravel(), y = previsoes)
grafico.show()

# a linha laranja representa o valor previsto
# os pontos azuis são os valores reais
```

```python
grafico1 = px.scatter(x = x_casas_treinamento.ravel(), y = y_casas_treinamento) # Respostas reais
grafico2 = px.line(x = x_casas_treinamento.ravel(), y = previsoes) # previsões
grafico2.data[0].line.color = 'red' # mudando a cor da linha de previsoes
grafico3 = go.Figure(data = grafico1.data + grafico2.data)
grafico3

# linha vermelha são as previsões
# bolinhas azuis são os valores reais
```

```python
# Comparando previões

previsoes_teste = regressor_simples_casas.predict(x_casas_teste)

previsoes_teste, y_casas_teste
```

```python
# Pegando a diferença dos valores absolutos

print(f'A média de erro é {abs(y_casas_teste - previsoes_teste).mean()}')
abs(y_casas_teste - previsoes_teste)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
# Diferenças reais

mean_absolute_error(y_casas_teste, previsoes_teste)
```

```python
# Valores elevado ao quadrado

mean_squared_error(y_casas_teste, previsoes_teste)
```

```python
np.sqrt(mean_squared_error(y_casas_teste, previsoes_teste))
```
