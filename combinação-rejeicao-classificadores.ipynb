# Combinação de classificadores

```python
# Bibliotecas utilizadas

import pickle
import numpy as np
```

```python
# Importando base de dados

with open('/content/drive/MyDrive/variáveis de teste e treinamento(credit, census)/credit.pkl', 'rb') as f:
  X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
# Concatenando as duas variáveis em uma só
# axis = 0, linhas.
# axis = 1, colunas.

X_credit = np.concatenate((X_credit_treinamento, X_credit_teste), axis = 0)
y_credit = np.concatenate((y_credit_treinamento, y_credit_teste), axis = 0)
```

```python
# Carregando algoritmo já treinado

rede_neural = pickle.load(open('/content/drive/MyDrive/classificadores treinados/rede_neural_finalizado.sav', 'rb'))
arvore = pickle.load(open('/content/drive/MyDrive/classificadores treinados/arvore_finalizado.sav', 'rb'))
svm = pickle.load(open('/content/drive/MyDrive/classificadores treinados/svm_finalizado.sav', 'rb'))
```

```python
# Pegando o registro da primeira linha do dataset

novo_registro = X_credit[1999]
novo_registro = novo_registro.reshape(1, -1)
```

```python
print(novo_registro)
print(novo_registro.shape)
```

```python
# Calculando resultados

resposta_rede_neural = rede_neural.predict(novo_registro)
resposta_arvore = arvore.predict(novo_registro)
resposta_svm = svm.predict(novo_registro)
```

```python
resposta_rede_neural[0], resposta_arvore[0], resposta_svm[0]
```

```python
paga = 0
nao_paga = 0

if resposta_rede_neural[0] == 1:
    nao_paga += 1
else:
    paga += 1

if resposta_arvore[0] == 1:
    nao_paga += 1
else:
    paga += 1

if resposta_svm[0] == 1:
    nao_paga += 1
else:
    paga += 1


if paga > nao_paga:
    print('O cliente pagará o empréstimo.')
elif paga == nao_paga:
    print('Empate.')
else:
    print('O cliente não pagará o empréstimo.')
```

# Rejeição de classificadores

```python
# Pegando a probabilidade

probabilidade_rede_neural = rede_neural.predict_proba(novo_registro)
probabilidade_arvore = arvore.predict_proba(novo_registro)
probabilidade_svm = svm.predict_proba(novo_registro)
```

```python
# Verificando a confiança

confianca_rede_neural = probabilidade_rede_neural.max()
confianca_arvore = probabilidade_arvore.max()
confianca_svm = probabilidade_svm.max()
```

```python
confianca_rede_neural, confianca_arvore, confianca_svm
```

```python
paga = 0
nao_paga = 0
confianca_minima = 0.999999
algoritmos = 0

if confianca_rede_neural >= confianca_minima:
  algoritmos += 1 # quer dizer que esse algoritmo foi usado
  if resposta_rede_neural[0] == 1:
    nao_paga += 1
  else:
    paga += 1

if confianca_arvore >= confianca_minima:
  algoritmos += 1 # quer dizer que esse algoritmo foi usado
  if resposta_arvore[0] == 1:
    nao_paga += 1
  else:
    paga += 1

if confianca_svm >= confianca_minima:
  algoritmos += 1 # quer dizer que esse algoritmo foi usado
  if resposta_svm[0] == 1:
    nao_paga += 1
  else:
    paga += 1

if paga > nao_paga:
  print(f'Cliente pagará o empréstimo, baseado em {algoritmos} algoritmos')
elif paga == nao_paga:
  print(f'Empate, baseado em {algoritmos} algoritmos')
else:
  print(f'Cliente não pagará o empréstimo, baseado em {algoritmos} algoritmos')
```
