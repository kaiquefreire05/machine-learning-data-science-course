# Naive Bayes base de crédito

```python
# Importando as variáveis do arquivo pkl
import pickle
with open('/content/drive/MyDrive/variáveis de teste e treinamento(credit, census)/credit.pkl', 'rb') as f:
    x_credit_treinamento,  y_credit_treinamento, x_credit_teste, y_credit_teste = pickle.load(f)
```

```python
# verificando tamanho das variáveis
print(x_credit_treinamento.shape, y_credit_treinamento.shape)
print(x_credit_teste.shape, y_credit_teste.shape)
```

```python
# Importando o algoritmo Gaussian do Naive Bayes
from sklearn.naive_bayes import GaussianNB
```

```python
# Criando a  variável construtora e fazendo a tabela de probabilidade
naive_credit_data = GaussianNB()
naive_credit_data.fit(x_credit_treinamento, y_credit_treinamento) # Usando apenas as variáveis de treinamento
```

```python
# Enviando para a variável um conjunto de dados que não está no database
previsoes = naive_credit_data.predict(x_credit_teste)
previsoes
```

```python
# Fazendo a comparação com as respostas reais
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
accuracy_score(y_credit_teste, previsoes)
```

```python
confusion_matrix(y_credit_teste, previsoes)
```

```python
print(classification_report(y_credit_teste, previsoes))
```

# Naive Bayes base do Census

```python
with open('/content/drive/MyDrive/variáveis de teste e treinamento(credit, census)/census.pkl', 'rb') as f:    # rb = Read(ler)
    x_census_treinamento, y_census_treinamento, x_census_teste, y_census_teste = pickle.load(f)
```

```python
print(x_census_treinamento.shape, y_census_treinamento.shape)
print(x_census_teste.shape, y_census_teste.shape)
```

```python
# Criando a  variável construtora e fazendo a tabela de probabilidade
naive_base_census = GaussianNB()
naive_base_census.fit(x_census_treinamento, y_census_treinamento) # Usando apenas as variáveis de treinamento
```

```python
# Enviando para a variável um conjunto de dados que não está no database
previsoes_census = naive_base_census.predict(x_census_teste)
previsoes_census
```

```python
# Comparando as variáveis reais com as previsões
print(f'A taxa de acerto é de: {accuracy_score(y_census_teste, previsoes_census)}') # Se não fazer o escalonamento a taxa pode aumentar para 70%
```

```python
print('Matriz de confusão: ')
print(confusion_matrix(y_census_teste, previsoes_census))
```

```python
print('Classificação: ')
print(classification_report(y_census_teste, previsoes_census))
```
