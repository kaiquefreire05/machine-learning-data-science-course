# Regressor usando Random Forests

# Base de planos de saúde

```python
# Importações

import numpy as np

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
import plotly.express as px
```

```python
# Importando as variáveis do arquivo pkl da base de plano de saúde

import pickle

with open('/content/drive/MyDrive/variaveis treino/saude.pkl', 'rb') as f:
    x_plano_saude, y_plano_saude = pickle.load(f)
```

```python
# Verificando o tamanho das variáveis

x_plano_saude.shape, y_plano_saude.shape
```

```python
# Criando o regressor e treinando

regressor_random_forest = RandomForestRegressor()
regressor_random_forest.fit(x_plano_saude, y_plano_saude)
```

```python
regressor_random_forest.score(x_plano_saude, y_plano_saude)
```

```python
# Criando valores aleatórios para fazer o teste no algoritmo

x_teste_arvore = np.arange(min(x_plano_saude), max(x_plano_saude), step=0.1)
x_teste_arvore.shape # a variável está no formato de vetor
```

```python
# Convertendo para matriz

x_teste_arvore = x_teste_arvore.reshape(-1, 1)
```

```python
# Gráfico para fazer a visualização dos splits

grafico = px.scatter(x = x_plano_saude.ravel(), y = y_plano_saude)
grafico.add_scatter(x = x_teste_arvore.ravel(), y = regressor_random_forest.predict(x_teste_arvore), name = 'Regressão')
grafico.show()
```

```python
# Teste de previsão

regressor_random_forest.predict([[40]])
```

# Base do preço das casas

```python
# Importando as variáveis da base de casas

with open('/content/drive/MyDrive/variaveis treino/casas.pkl', 'rb') as f:
    x_casas_treinamento, y_casas_treinamento, x_casas_teste, y_casas_teste = pickle.load(f)
```

```python
# Verificando o tamanho das variáveis

print(f'O tamanho das variáveis de treinamento é: {x_casas_treinamento.shape, y_casas_treinamento.shape}')
print(f'O tamanho das variáveis de teste é: {x_casas_teste.shape, y_casas_teste.shape}')
```

```python
# Criando o regressor e fazendo o treinamento

regressor_random_casas = RandomForestRegressor()
regressor_random_casas.fit(x_casas_treinamento, y_casas_treinamento)
```

```python
# Visualizando o score com as variáveis de treino

regressor_random_casas.score(x_casas_treinamento, y_casas_treinamento)
```

```python
# Visualizando o score com as variáveis de teste

regressor_random_casas.score(x_casas_teste, y_casas_teste)
```

```python
# Fazendo previsoes

previsoes = regressor_random_casas.predict(x_casas_teste)
```

```python
# Média de erro absoluto

mean_absolute_error(y_casas_teste, previsoes)
```
