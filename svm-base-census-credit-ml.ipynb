```python
# Biblioteca para usar o algoritmo
from sklearn.svm import SVC
```

# Base Credit Data

```python
import pickle
with open('/content/drive/MyDrive/variáveis de teste e treinamento(credit, census)/credit.pkl', 'rb') as f:
    x_credit_treinamento, y_credit_treinamento, x_credit_teste, y_credit_teste = pickle.load(f)
```

```python
print(x_credit_treinamento.shape, y_credit_treinamento.shape)
print(x_credit_teste.shape, y_credit_teste.shape)
```

```python
# Criando algortimo e fazendo o treino
svm_credit = SVC(kernel='rbf', random_state=1, C=2.0) # Esses parâmetros podem ser mudados
svm_credit.fit(x_credit_treinamento, y_credit_treinamento)

# Kernel 'linear': 0.946
# Kernel 'poly' (polinominal): 0.968
# Kernel 'sigmoid'(sigmoide): 0.838
# Kernel 'rbf' (default, tendência em ser melhor na maioria de base de dados): 0.982

# Com kernel 'rbf' que foi melhor r C=1.0: 0.982
# Com kernel 'rbf' que foi melhor r C=2.0: 0.988
# Com kernel 'rbf' que foi melhor r C=3.0: 0.984
```

```python
# Fazendo as previsões
previsoes_credit = svm_credit.predict(x_credit_teste)
previsoes_credit
```

```python
# Importando as métricas
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
```

```python
# Fazendo as métricas
print(f'A taxa de acurácia do algoritmo foi de {accuracy_score(y_credit_teste, previsoes_credit)}')
print('=-='*20)
print(f'\nRelatório de classificação: \n{classification_report(y_credit_teste, previsoes_credit)}')
print('=-='*20)
print('\nMatriz de confusão: ')
cm = ConfusionMatrix(svm_credit)
cm.fit(x_credit_treinamento, y_credit_treinamento)
cm.score(x_credit_teste, y_credit_teste)
```
