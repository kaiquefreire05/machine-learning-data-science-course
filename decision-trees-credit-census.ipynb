```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Pegando minhas variáveis usando o pickle
import pickle
with open('/content/drive/MyDrive/variáveis de teste e treinamento(credit, census)/risco_credito.pkl', 'rb') as f:
    x_risco_credito, y_risco_credito = pickle.load(f)
```

# Base de crédito - 98.20%

```python
# Importando as variáveis
with open('/content/drive/MyDrive/variáveis de teste e treinamento(credit, census)/credit.pkl', 'rb') as f:
    x_credit_treinamento, y_credit_treinamento, x_credit_teste, y_credit_teste = pickle.load(f)
```

```python
# Visulizando o tamanho das variáveis de treino e teste
print(x_credit_treinamento.shape, y_credit_treinamento.shape)
print(x_credit_teste.shape, y_credit_teste.shape)
```

```python
# Criando a variável construtora
arvore_credit = DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
# Passando as variáveis para ser criada a árvore de decisão
arvore_credit.fit(x_credit_treinamento, y_credit_treinamento)
```

```python
# Visualizando os atributos mais importantes
arvore_credit.feature_importances_
```

```python
# Classes(resultado)
arvore_credit.classes_
```

```python
# Fazendo as previsões
previsoes_credit = arvore_credit.predict(x_credit_teste)
previsoes_credit
```

```python
# Vendo a porcentagem de acerto
from sklearn.metrics import accuracy_score, classification_report
print(f'A porcentagem de acerto é de: {accuracy_score(y_credit_teste, previsoes_credit)}')
```

```python
# Gerando a matriz de confusão
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(arvore_credit)
cm.fit(x_credit_treinamento, y_credit_treinamento)
cm.score(x_credit_teste, y_credit_teste)
```

```python
# Criando uma tabela de classificação
print(classification_report(y_credit_teste, previsoes_credit))
```
